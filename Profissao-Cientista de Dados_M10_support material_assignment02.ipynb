````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
````

1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
df_renda = pd.read_csv('previsao_de_renda.csv')
df_renda = df_renda.drop(columns='Unnamed: 0').set_index(keys='index', drop=True)
df_renda.drop_duplicates(inplace=True)
df_renda
````

````python
plt.rc('figure', figsize=(12,12))
fig, axes = plt.subplots(2,2)

sns.pointplot(x='posse_de_imovel', 
              y='renda',  
              data=df_renda, 
              dodge=True, 
              ax=axes[0,0])

sns.pointplot(x='posse_de_veiculo', 
              y='renda', 
              data=df_renda, 
              dodge=True, 
              ax=axes[0,1])

sns.pointplot(x='posse_de_imovel', 
              y='renda',  
              hue='posse_de_veiculo', 
              data=df_renda, 
              dodge=True, 
              ax=axes[1,0])

sns.pointplot(x='posse_de_veiculo', 
              y='renda', 
              hue='posse_de_imovel', 
              data=df_renda, 
              dodge=True, 
              ax=axes[1,1])

plt.show()
````

A posse de veículo parece ser um melhor preditor de renda do que a posse de imóvel, pois a diferença de renda entre os grupos é mais clara e os intervalos de confiança sugerem uma separação estatisticamente mais significativa. Isso pode indicar que pessoas com veículos possuem empregos que demandam transporte próprio ou têm maior capacidade financeira para aquisição de bens.
